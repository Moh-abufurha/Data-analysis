```python
counts = tips.groupby(['day', 'sex']).size().unstack()
counts.plot(kind='bar', color=[ 'blue','red'])
plt.title('Number of Meals by Day and Sex')
```

```python
tim=tips.groupby(['time', 'sex']).size().unstack()

tim.plot(kind='pie', colors=['black', 'red'], autopct='%1.1f%%', subplots=True)
plt.title('Number of sex by and time')
plt.figure(figsize=(10, 10))
```

```python
tim=tips.groupby(['smoker', 'sex']).size().unstack()

tim.plot(kind='pie', colors=['gold', 'olivedrab'], autopct='%1.1f%%', subplots=True)
plt.title('Number of somker by and Sex')
plt.figure(figsize=(10, 10))
```

```python
tim=tips.groupby(['size', 'sex']).size().unstack()

tim.plot(kind='pie', colors=['royalblue', 'darkorchid','blue','mediumpurple','navy','plum'], autopct='%1.1f%%', subplots=True)
plt.title('size by and Sex')
plt.figure(figsize=(10, 45))
```

```python
x=tips.groupby(['size','smoker'])['tip'].mean().unstack()
x.plot(kind='bar',color=['salmon','springgreen'])

plt.title('size by and smoker')
```
